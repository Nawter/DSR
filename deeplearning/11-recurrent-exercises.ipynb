```
from keras.models import Sequential
from keras import layers

model = Sequential()
# connect LSTM with LSTM return_sequences=True
model.add(layers.LSTM(32, activation='relu',return_sequences=True, input_shape=(10, 28*28)))
model.add(layers.LSTM(16, activation='relu',return_sequences=True ))
model.add(layers.LSTM(16, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))
model.summary()
```
