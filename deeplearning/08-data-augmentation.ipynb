```python
from keras.preprocessing.image import ImageDataGenerator
# rescale the colors
datagen = ImageDataGenerator(
    rotation_range = 40,
    width_shift_range = 0.2,
    height_shift_range = 0.2,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True,
    fill_mode = "nearest"    
)
```

```python
%matplotlib inline
from keras.preprocessing import image
import matplotlib.pyplot as plt
import numpy as np
import glob


fnames = glob.glob("cats_vs_dogs_small/train/**/*.jpg", recursive=True)
#fnames  = [fname for fname in fnames if "dog." in fname]
print(fnames[2])

img = image.load_img(fnames[2], target_size=(150,150))


x = image.img_to_array(img)
x = np.reshape(x,(1,) + x.shape)
print(x.shape)



#import os

#fnames = [os.path.join(train_cats_dir, fname)
#        for fname in os.listdir(train_cats_dir)]
# train_generator is equivalent to datagen.flow(x, batch_size=1)
i = 0
for batch in datagen.flow(x, batch_size=1):
    plt.figure(i)
    imgplot = plt.imshow(image.array_to_img(batch[0]))
    i +=1
    if i% 4 == 0:
        break
```
